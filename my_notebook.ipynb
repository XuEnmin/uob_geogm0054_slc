```python
import numpy as np
import matplotlib.pyplot as plt
from calendar import month_name
```

```python
year, month, tmax, tmin = np.loadtxt('rossonwyedata.txt',skiprows=7,usecols=(0,1,2,3),unpack=True)
```

```python
month_nums = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'] # 1 to 12, inclusive
month_names = [month_nums[mm]  for mm in range(0,12)]

#create a list of 1D arrays, 1 for each month
tmax_by_month = [tmax[month == mm] for mm in range(0,12)]

#HINT reverse the month order to place January at the top 123456
#tmax_by_month.reverse() 112233
#month_names.reverse()

#boxplot produces one box-and whisker for each 1D array in a list
fig, ax = plt.subplots(1,1)
__ = ax.boxplot(tmax_by_month, vert=False,tick_labels=month_names)
# HINT add the keyword argument labels=month_names (Python < 3.9) or tick_labels=month_names (Python >= 3.9)
__ = ax.set_xlabel(r'Temperature ($^\circ$C)')
__ = ax.set_ylabel('Month')
__ = ax.set_title(f'Temperature, {int(np.min(year))} to {int(np.max(year))}')
```
